```python
(ggplot(mpg)         
 + aes(x='class')    
 + geom_bar(size=20)
 + labs(title="Testing", subtitle="Subtitle testing")
 + scale_y_continuous(expand=(0,0)) # start at 0 with no gap
 + xlab("")
 + ylab("")
 + theme_ic
)
```

```python
cbp_data = pd.read_csv('data/cbp_data_2024-04-08.csv')
cbp_data['FY'] = cbp_data['FY'].str.replace(' (FYTD)', '')
cbp_data['date'] = pd.to_datetime(cbp_data['FY'] + '-' + cbp_data['Month (abbv)'] + '-' + '01',
                                  format="%Y-%b-%d")

sw_data = cbp_data[cbp_data['Region'] == 'Southwest Border']
```

```python
(ggplot(sw_data)
 + aes(x='date', y='Sum Qty (lbs)', color='Drug Type')
 + geom_line()
 + scale_y_continuous(expand=(0,0))
 + theme_ic
 + facet_wrap('Area of Responsibility')
)
```
